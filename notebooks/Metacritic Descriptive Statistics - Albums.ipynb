```python
import pandas as pd
import numpy as np
```

```python
albums_df = pd.read_json("../output/albums.json")
albums_df
```

```python
albums_df.describe().round(1)
```

```python
import plotly
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

plotly.offline.init_notebook_mode(connected=True)
```

```python
data1 = go.Histogram(
    x = albums_df['usr rating'].multiply(10),
    opacity=0.75,
    name="user ratings"
)
data1_1 = go.Histogram(
    x = albums_df['mc rating'],
    opacity=0.75,
    name="metacritic ratings"
)

layout1 = go.Layout(
    title= 'Histogram of Album Scores',
    #hovermode= 'closest',
    xaxis = dict(
        title= 'Score'
    ),
    yaxis = dict(
        title = 'Number of Ratings'
    )
)

fig1 = dict(data=[data1,data1_1], layout=layout1)
iplot(fig1, filename='basic histogram')
```
