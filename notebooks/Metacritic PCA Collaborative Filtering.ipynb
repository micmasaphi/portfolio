```python
# reference from https://beckernick.github.io/matrix-factorization-recommender/

import pandas as pd
import numpy as np
```

```python
# read from PostgreSQL once db is available
ratings_df = pd.read_csv("data/ratings.csv")
ratings_df.head()
```

```python
# TODO: fill_value = 0?
ratings_df = pd.pivot_table(ratings_df, index='critic_id', columns='album_id', values='album_rating',aggfunc=np.max,fill_value=0);
# ratings_df.to_csv('data/review_matrix.csv', na_rep="")
ratings_df.head()
```

```python
R = ratings_df.as_matrix()
ratings_df.shape
```

```python
# TODO: consider axis = 0 for analysis?
# TODO: .mean vs .nanmean?
critic_ratings_mean = np.nanmean(R, axis = 1) # axis = 1 means we take the average of each row (critics in this case)
R_demeaned = R - critic_ratings_mean.reshape(-1,1)
```

```python
# benchmarking to consider (ARPACK seems slow; not sure how often this will be run; mostly offlined, no?)
# https://simplyml.com/benchmarking-the-singular-value-decomposition/
from scipy.sparse.linalg import svds
```

```python
# TODO: consider sklearn TruncatedSVD? https://stackoverflow.com/questions/18706863/difference-between-truncatedsvd-and-svds
"""
***SVD***
R = U*(Sigma)*Vt
R = user ratings matrix
U = user features matrix
Sigma = diagonal matrix of signular values (weights)
Vt = movie features matrix
"""
U, sigma, Vt = svds(R_demeaned) # k = 6 is default
```
